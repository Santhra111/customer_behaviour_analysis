```python
# we need to fill the entire rating with median of overall rating, because it is robust to outliers than mean
```

```python
# but it is not a good practice clothing rating can be entirely different from footwear category and entirely different from bag category
```

```python
# what we are gonna do is if a rating item clothing item get missed we replace with it clothing item not every item
```

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
#transform() performs a group-wise operation but keeps the original row structure.
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
 #create a column Age group
```

```python
# ðŸ”µ Why Do We Use qcut()?

# ðŸ‘‰ To divide data into equal-sized groups.

# Thatâ€™s it.

# ðŸŽ¯ Real Reason in Data Analysis

# Sometimes companies donâ€™t care about exact age ranges.

# They care about:

# Top 25% customers

# Bottom 25% customers

# Middle customers

# So we divide people equally.
```

```python
labels=['Youngadult','Adult','Middle-Aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#lables-  Use these names for the groups instead of numeric ranges
```

```python
#Create Columns purchase_frequency_days
```

```python
# Mapping purchase frequency categories to equivalent number of days
# Models cannot understand text like "Weekly"

# Numeric values are easier for analysis

# Helps in calculations (average purchase gap, etc.)
```

```python
Frequency_mapping={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly': 30,
    'Quarterly':90,
    'Bi-weekly':14,
    'Annually': 365,
    'Every 3 months': 90
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(Frequency_mapping)
    
```
